```python
import pandas as pd 

import os
from bokeh.plotting import figure, ColumnDataSource
from bokeh.models.widgets import Dropdown
from bokeh.io import curdoc,show, output_notebook
from bokeh.layouts import column

from bokeh.models import Select, Range1d, HoverTool
from bokeh.palettes import Category20
from bokeh.models.formatters import NumeralTickFormatter

output_notebook()
```

```python
# Read data
df=pd.read_excel('fb.xlsx')
df.reset_index(inplace=True)
#df['Date']=pd.to_datetime(df['Date'])
df['tooltip'] = [x.strftime("%Y-%m-%d") for x in df['Date']]
df.head()
```

```python
from bokeh.models import DatetimeTickFormatter
src = ColumnDataSource(df)
p = figure(plot_height = 600, plot_width = 600, 
           title = 'FB Stock Price',
          x_axis_label = 'Dates', 
           y_axis_label = 'Close Price',
          x_axis_type='datetime')

#p.xaxis.formatter=DatetimeTickFormatter(
 #       hours=["%d %B %Y"],
  #      days=["%d %B %Y"],
   #     months=["%d %B %Y"],
    #    years=["%d %B %Y"],
    #)

p.xaxis.major_label_orientation = 3.14/4

# Add a line glyph
p.line(source=src, x='Date', y='Close', 
         color='red', line_width=1,
         legend='Close Price')

p.legend.location="top_left"

# Show the plot
show(p)
```

```python
def style(p):
        # Title 
        p.title.align = 'center'
        p.title.text_font_size = '20pt'
        p.title.text_font = 'serif'

        # Axis titles
        p.xaxis.axis_label_text_font_size = '14pt'
        p.xaxis.axis_label_text_font_style = 'bold'
        p.yaxis.axis_label_text_font_size = '14pt'
        p.yaxis.axis_label_text_font_style = 'bold'

        # Tick labels
        p.xaxis.major_label_text_font_size = '12pt'
        p.yaxis.major_label_text_font_size = '12pt'

        return p
```

```python
hover = HoverTool(tooltips = [('Close price', '@Close'),
                             ('Dates', '@tooltip')])
```

```python
p.add_tools(hover)

#p=style(p)
# Show the plot
show(p)
```
